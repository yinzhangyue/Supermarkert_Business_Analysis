## 探索性数据分析

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('./supermarket_sales.csv')
```

```python
sales.head()
```

```python
sales.info()
```

```python
sales['date'] = pd.to_datetime(sales['Date'])
```

```python
sales['day'] = (sales['date']).dt.day
sales['month'] = (sales['date']).dt.month
sales['year'] = (sales['date']).dt.year
```

```python
sales['Time_'] = pd.to_datetime(sales['Time'])
sales['Hour'] = (sales['Time_']).dt.hour
sales['Hour'].unique()
```

```python
sales.head()
```

```python
sales.columns
```

```python
longitude = {"Yangon": 16.8661, "Naypyitaw": 19.7633, "Mandalay": 21.9588 }
latitude = {"Yangon": 96.1951, "Naypyitaw": 96.0785, "Mandalay": 96.0891 }
for set in sales:
    sales['longitude'] = sales['City'].map(longitude)
    sales['latitude'] = sales['City'].map(latitude)
```

```python
city = {"Yangon": '仰光', "Naypyitaw": '内比都', "Mandalay": '曼德勒'}
def t1(input):
    return city[input]
sales['City'] = sales['City'].apply(t1)

customertype = {"Member": '会员', "Normal": '非会员'}
def t2(input):
    return customertype[input]
sales['Customer type'] = sales['Customer type'].apply(t2)

gender = {"Female": '女性', "Male": '男性'}
def t3(input):
    return gender[input]
sales['Gender'] = sales['Gender'].apply(t3)

productline = {"Health and beauty": '健康与美容', 
               "Electronic accessories": '电子配件', 
               "Home and lifestyle": '生活家居', 
               "Sports and travel": '运动与旅行', 
               "Food and beverages": '饮食', 
               "Fashion accessories": '时尚配饰'}
def t4(input):
    return productline[input]
sales['Product line'] = sales['Product line'].apply(t4)

payment = {"Ewallet": '电子钱包', "Cash": '现金', "Credit card": '信用卡'}
def t5(input):
    return payment[input]
sales['Payment'] = sales['Payment'].apply(t5)
```

```python
del sales['Date']
del sales['Time_']
del sales['gross margin percentage']
sales.columns
```

```python
import scipy.stats
import numpy as np
A = np.array([60,47,65,51,58,59])
B = np.array([55,53,50,62,50,62])
C = np.array([55,52,45,65,66,45])
scipy.stats.kruskal(A,B,C)
```

```python
Q_A = pd.DataFrame(sales[sales['超市'] == 'A'].groupby(['交易月','交易日'])['毛利'].sum()).reset_index()
Q_A['A'] = 1
```

```python
Q_B = pd.DataFrame(sales[sales['超市'] == 'B'].groupby(['交易月','交易日'])['毛利'].sum()).reset_index()
Q_B['B'] = 1
```

```python
Q_C = pd.DataFrame(sales[sales['超市'] == 'C'].groupby(['交易月','交易日'])['毛利'].sum()).reset_index()
```

```python
Q_AB = pd.merge(Q_A, Q_B, on=['交易月','交易日'], how='left')
Q_AB['毛利_y'] = Q_AB['毛利_y'].fillna(0)
```

```python
Q_ABC = pd.merge(Q_AB, Q_C, on=['交易月','交易日'], how='left')
Q_ABC['毛利'] = Q_ABC['毛利'].fillna(0)
```
